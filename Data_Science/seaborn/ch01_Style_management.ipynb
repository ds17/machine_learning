## 1- Controlling figure aesthetics
语句 | 功能
:----|:----
`sns.set_style("ticks")` | 设置样式
`sns.boxplot(data=data)`| 箱形图
`sns.despine()`| 去除边框
`sns.despine(offset=20,trim=True)`| 边框偏移并修剪
`with sns.axes_style("darkgrid"):`| 设置局部样式
`sns.axes_style()`| 查看局部样式的参数
`sns.despine(left=True)` | 
`sns.set_style("darkgrid",{"axes.facecolor":"0.7"})`|
`sns.set_context("paper")` |

```python
sns.axes_style()
```

```python
sns.set(style='ticks')
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
sns.set_style("whitegrid")
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.boxplot(data=data);
```

```python
sinplot()
sns.despine(offset=20,trim=True)
```

```python
sinplot()
sns.despine(left=True)
```

```python
with sns.axes_style("darkgrid"):
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(-1)
```

```python
sns.set_style("darkgrid",{"axes.facecolor":"0.7"})
sinplot()
```

```python
sns.set_style("darkgrid",{"axes.facecolor":"0.1"})
sinplot()
```

```python
sns.set_context("paper")
sinplot()
```

## 2 - Color Palettes

### `color_palette()`
- defaut: `deep`
- `deep, muted, pastel, bright, dark,` and `colorblind`

需要的颜色 > 调色板中颜色数量时，开始循环使用

为避免颜色重复不好区分，使用 `hls` 调色板

```python
sns.palplot(sns.color_palette())
```

```python
sns.palplot(sns.color_palette("colorblind",10))
```

### `hsl_palette()`
### `husl_palette()`
- `hsl_palette(num, l=?, s=?)`
- `husl_palette(num, l=?, s=?)`
- HSL色彩模式
    - `H` - Hue，色相
    - `S` - Saturation，饱和度
    - `L` - Lighteness，明度

```python
sns.palplot(sns.color_palette("hls",20))
```

```python
sns.palplot(sns.hls_palette(10,l=0.6, s=0.9))
```

```python
sns.palplot(sns.husl_palette(10,l=0.6))
```

```python
sns.palplot(sns.color_palette("Paired",12))
```

```python
sns.palplot(sns.color_palette("Set2",12))
```

```python
flatui=["#9b59b6", "#3498db", "#95a5a6", "#e74c3c", "#34495e", "#2ecc71"]
sns.palplot(sns.color_palette(flatui,8))
```

### `xkcd_rgb["color name"]`
[color name index](https://xkcd.com/color/rgb/)
- `sns.xkcd_rgb["name"]`
- `sns.xkcd_palette(["color name list"])`

```python
plt.plot([0,1],[0,1], sns.xkcd_rgb["pale red"], lw=2)   #lw:lineweight
plt.plot([0,1], [0,2], sns.xkcd_rgb["medium green"], lw=2)
plt.plot([0,1], [0,3], sns.xkcd_rgb["dusty purple"], lw=2);
```

```python
colors=["windows blue", "amber", "greyish", "faded green", "dusty purple"]
sns.palplot(sns.xkcd_palette(colors))
```
